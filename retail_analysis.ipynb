```python
#import csv fils
retail =pd.read_csv("SampleSuperstore.csv")
retail.head(5)
```

```python
sns.scatterplot(x="Profit",y="Region", data=retail)
plt.show()
```

```python
sns.barplot(x="Category",y="Profit" ,data=retail)
plt.show()
```

```python
sns.barplot(x="Sub-Category" ,y="Discount",data=retail)
plt.show()
```
